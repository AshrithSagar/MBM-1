## Upload input files

```python
jobname = 'MAM7_V1_12_2_4' #@param {type:"string"}
```

```python
# @title Upload **rec.pdb**

import os
from google.colab import files

rec = files.upload()

for fn in rec.keys():
  os.rename(fn, f"{jobname}_rec.pdb")
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(rec[fn])))
```

```python
# @title Upload **lig.pdb**

import os
from google.colab import files

lig = files.upload()

for fn in lig.keys():
  os.rename(fn, f"{jobname}_lig.pdb")
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(lig[fn])))
```

## Run *piper* Job

```python
# @title Run PIPER Job
%%bash -s $jobname $rec $lig

jobname=$1
rec=$2
lig=$3

!python piper_package/protein_prep/prepare.py "${jobname}_${rec}.pdb"

!python piper_package/protein_prep/prepare.py "${jobname}_${lig}.pdb"

!python piper_package/piper -vv -c1.0 -k4 --msur_k=1.0 --maskr=1.0 -T FFTW_EXHAUSTIVE -R 70000 -t 1 -p piper_package/prms/atoms.prm -f piper_package/prms/coeffs.prm -r piper_package/prms/rots.prm "${rec}_pnon.pdb" "${lig}_pnon.pdb"
```
